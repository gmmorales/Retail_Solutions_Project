# Generador de Datos Sintéticos - Retail Solutions X
**Proyecto:** Diagnóstico de Caída de Ventas y Comportamiento de Clientes  
**Descripción:** Este notebook genera el dataset inicial de 50,000 registros para simular el entorno de consultoría.  
**Archivo de salida:** `../data/raw/ventas_consolidadas_2024_2025.csv`

```python
# 1. Asegurar que las librerías necesarias estén instaladas en este entorno
# El símbolo '%' asegura que se instale en el kernel actual: 'Retail Solutions X'
%pip install pandas numpy

# 2. Importaciones fundamentales
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta

print("✅ Librerías importadas y listas para generar datos.")
```

```python
# 1. Configuración de ruta
folder_path = '../data/raw/'
if not os.path.exists(folder_path):
    os.makedirs(folder_path, exist_ok=True)

# 2. Parámetros del dataset
n_rows = 50000
np.random.seed(42)

# Listas de categorías y métodos
categorias = ['Electro', 'Hogar', 'Moda', 'Audio', 'Deportes']
metodos_pago = ['Tarjeta de Crédito', 'Transferencia', 'Efectivo', 'Billetera Digital']

# 3. Generación de datos sintéticos
print(f"Generando {n_rows} registros... Por favor espera un momento.")

data = {
    'id_transaccion': range(100001, 100001 + n_rows),
    'id_cliente': np.random.randint(10000, 15000, size=n_rows), # Simula 5000 clientes únicos
    'fecha_compra': [(datetime(2024, 1, 1) + timedelta(days=np.random.randint(0, 730))).strftime('%Y-%m-%d') for _ in range(n_rows)],
    'categoria_prod': [np.random.choice(categorias) for _ in range(n_rows)],
    'precio_bruto': np.round(np.random.gamma(shape=2, scale=200, size=n_rows) + 50, 2), # Distribución más realista
    'descuento_aplicado': np.round(np.random.choice([0.0, 0.05, 0.10, 0.20], size=n_rows, p=[0.6, 0.2, 0.1, 0.1]), 2),
    'metodo_pago': [np.random.choice(metodos_pago) for _ in range(n_rows)],
    # Introducimos 5% de nulos en edad para que practiques la limpieza
    'edad_cliente': np.random.choice([np.nan, 22, 28, 35, 42, 55, 65], size=n_rows, p=[0.05, 0.15, 0.2, 0.25, 0.15, 0.1, 0.1])
}

# 4. Crear DataFrame
df_final = pd.DataFrame(data)

# Ordenar por fecha para que sea más natural
df_final['fecha_compra'] = pd.to_datetime(df_final['fecha_compra'])
df_final = df_final.sort_values('fecha_compra').reset_index(drop=True)

# 5. Guardar como CSV
file_path = os.path.join(folder_path, 'ventas_consolidadas_2024_2025.csv')
df_final.to_csv(file_path, index=False)

print(f"--- PROCESO COMPLETADO ---")
print(f"Archivo creado en: {file_path}")
print(f"Total de registros: {len(df_final)}")
print(f"Memoria utilizada: {df_final.memory_usage().sum() / 1024**2:.2f} MB")
print("\nPrimeras 5 filas:")
print(df_final.head())
```
